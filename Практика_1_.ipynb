## 4. Визуализация распределения переменных

1. **Гистограмма для числовой переменной age:**

```python
sns.histplot(df['age'], kde=True)
plt.title('Распределение age')
plt.xlabel('age')
plt.ylabel('Частота')
plt.show()
```

2. **Столбчатая диаграмма для категориальной переменной chest pain type:**

```python
sns.countplot(x='chest pain type', data=df)
plt.title('Распределение chest pain type')
plt.xlabel('chest pain type')
plt.ylabel('Количество')
plt.show()
```

**Боксплот для числовой переменной resting blood pressure:**

```python
sns.boxplot(x=df['resting blood pressure'])
plt.title('Boxplot resting blood pressure')
plt.xlabel('resting blood pressure')
plt.show()
```

**Наблюдения:**

Большинство людей в выборке от 40 до 70 лет с пиком в возрасте 60 лет. Типы болей в груди расположены по частоте встречаемости от первого до четвертого соответственно и каждый последующий тип встречается в два раза чаще предыдущего.Минимальное кровяное давление меньше 100, 25% приходится на 120, 50% на 130, 75% на 140 и максимальное на 170. Выше 170 это исключения.
